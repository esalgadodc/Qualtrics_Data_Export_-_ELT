```python
import pandas as pd

df = pd.read_csv("Doc.csv")
```

```python
df.head()
```

```python
df = df.drop(columns=['LocationLatitude', 'LocationLongitude', 'Question 5.B','Question 6_1', 'Question 6_2', 'Question 6_3'])
df.head()
```

```python
df = df.rename(columns={
    'Question 1' : 'Question 1 - How often do you drink Matcha Tea?',
    'Question 2' : 'Question 2 - On average, how many cups of Matcha do you drink per day?',
    'Question 3' : 'Question 3 - How many days a week do you typically drink matcha tea?',
    'Question 4' : 'Question 4 - What is your favorite way to enjoy Matcha?', 
    'Question 4B' : 'Question 4.B - If other, please describe',
    'Question 5' : 'Question 5 - What flavors would you be most inclined to try in your Matcha beverage? (Select Top 3)'
})

print(df.columns)
df.head()
```

```python
df = df.drop(index=[0, 1])
df.head()
```

```python
df.to_excel("Qualtrics_Matcha_ETL_ES.V11.22.xlsx", index=False)
```

```python
df_split = df['Question 5 - What flavors would you be most inclined to try in your Matcha beverage? (Select Top 3)'].str.split(',', expand=True)

df_split.columns = ['Matcha Flavor Choice 1', 'Matcha Flavor Choice 2', 'Matcha Flavor Choice 3']

df_final = pd.concat([df, df_split], axis=1)

all_choices = df_split.columns.dropna()

print(df_final)
df_final.head()
```

```python
df.to_csv("Qualtrics_Matcha_ETL_ES.V11.22.GH.csv", index=False)
```

```python
df_final.head()
```

```python
df_final.to_excel("Qualtrics_Matcha_ETL_ES.V11.22.GH_Final_Post.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
all_matcha_choices =  df_final['Question 5 - What flavors would you be most inclined to try in your Matcha beverage? (Select Top 3)'].str.split(',').explode()
choice_counts = all_matcha_choices.value_counts()
plt.figure(figsize=(8, 6))
choice_counts.plot(kind='bar', color='green')
plt.title('Survey Responses: Favorite Matcha Flavors')
plt.xlabel('Flavors')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
matcha_type =  df_final['Question 4 - What is your favorite way to enjoy Matcha?'].str.split(',').explode()
choice_counts = matcha_type.value_counts()
plt.figure(figsize=(8, 6))
choice_counts.plot(kind='bar', color='green')
plt.title('Survey Responses: Types of Ways to Have Matcha')
plt.xlabel('Ways to have Matcha')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.show()
```
